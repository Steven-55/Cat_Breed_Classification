```python
from tensorflow import keras

# Load the H5 model
# '/kaggle/input/modelcatbreed/model.h5' Path dataset
model = keras.models.load_model('/kaggle/input/modelcatbreed/model.h5')
```

```python
import cv2 #import library
import numpy as np
from tensorflow.keras.preprocessing import image
import time

image_size = (224, 224) #Semua input di convert ke size ini

# Load Image
image_path = '/kaggle/input/bengal/20210708152424-1-kucing-bengal-dan-4-ras-lainnya-dari-amerika-kebanggaan-negara-paman-sam-002-jeffreydien-winanda.jpg'  # Image Path
img = cv2.imread(image_path) #load image
img = cv2.resize(img, image_size) #resize
img = img.astype('float32') / 255.0 #Convert dan normalize sehingga menjadi antara 0-1
img = np.expand_dims(img, axis=0) # tambah 1 dimensi

predictions = model.predict(img) #ngepredict dengan model

```

```python

class_labels = ["Bengal", "Bombay", "Persian", "Torbie", "Tuxedo"] #Ini untuk label output

for i, prediction in enumerate(predictions[0]): #Iterasi ke semua hasil prediksi
    class_label = class_labels[i] 
    confidence = prediction * 100 #Sehingga bentuk persen
    print("Class:", class_label)
    print("Confidence:", confidence, "%")

predicted_class_index = np.argmax(predictions[0]) #Urutkan ke terbesar untuk dapat hasil terbaik
predicted_class_label = class_labels[predicted_class_index]
highest_confidence = predictions[0][predicted_class_index] * 100
```

```python
# Print tertinggi
print("")
print("The Cat Is", predicted_class_label, "With Highest Confidence:", highest_confidence, "%")
```
